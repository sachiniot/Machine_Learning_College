```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.pipeline import make_pipeline
```

```python
df=pd.read_csv("C:\\Users\\mukes\\OneDrive\\Documents\\ML\\ipl data set.csv")
df
```

```python
df.columns
```

```python
x=df[['batting_team','bowling_team','venue','pitch_condition','current_score','overs_completed','wickets_out','runs_last_5_overs','overs_remaining','wickets_remaining','run_rate']]
y=df["final_score"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
ordinal_order = [["Green", "Dusty", "Dry", "Hard", "Flat"]]
```

```python
preprocessor = ColumnTransformer(transformers=[
    
    ('tnf1',OrdinalEncoder(categories=ordinal_order),['pitch_condition']),
    ('tnf2',OneHotEncoder(sparse_output=False,drop='first'),['batting_team','bowling_team','venue'])
],remainder='passthrough')
```

```python
# Transform
x_train_transformed = preprocessor.fit_transform(x_train)
x_test_transformed = preprocessor.transform(x_test)

# Convert back to DataFrame
x_train_transformed = pd.DataFrame(
    x_train_transformed,
    columns=preprocessor.get_feature_names_out(),
    index=x_train.index
)

x_test_transformed = pd.DataFrame(
    x_test_transformed,
    columns=preprocessor.get_feature_names_out(),
    index=x_test.index
)
```

```python
x_train_transformed
```

```python
model1=LinearRegression()
model1.fit(x_train_transformed,y_train)
y_predict=model1.predict(x_test_transformed)
print(r2_score(y_test,y_predict))
```

```python
model2=make_pipeline(PolynomialFeatures(degree=2),LinearRegression())
model2.fit(x_train_transformed,y_train)
y_predict2=model2.predict(x_test_transformed)
print(r2_score(y_test,y_predict2))
```

```python
model3=make_pipeline(PolynomialFeatures(degree=3),LinearRegression())
model3.fit(x_train_transformed,y_train)
y_predict3=model3.predict(x_test_transformed)
print(r2_score(y_test,y_predict3))
```
