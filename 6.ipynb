```python
"""  Making models for predicing the final score
of an inning . """
```

```python
x=df[['batting_team', 'bowling_team', 'venue',
       'pitch_condition', 'current_score', 'overs_completed', 'wickets_out',
       'runs_last_5_overs', 'overs_remaining', 'wickets_remaining', 'run_rate']]
y=df["final_score"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
ordinal_order = [["Green", "Dusty", "Dry", "Hard", "Flat"]]
```

```python
preprocessor1= ColumnTransformer(transformers=[
    
    ('tnf1',OrdinalEncoder(categories=ordinal_order),['pitch_condition']),
    ('tnf2',OneHotEncoder(sparse_output=False,drop='first'),['batting_team','bowling_team','venue'])
],remainder='passthrough')
```

```python
# Transform
x_train_transformed = preprocessor1.fit_transform(x_train)
x_test_transformed = preprocessor1.transform(x_test)

# Convert back to DataFrame
x_train_transformed = pd.DataFrame(
    x_train_transformed,
    columns=preprocessor1.get_feature_names_out(),
    index=x_train.index
)

x_test_transformed = pd.DataFrame(
    x_test_transformed,
    columns=preprocessor1.get_feature_names_out(),
    index=x_test.index
)
```

```python
x_train_transformed
```

```python
model1=LinearRegression()
model1.fit(x_train_transformed,y_train)
y_predict=model1.predict(x_test_transformed)
print(r2_score(y_test,y_predict))
```

```python
model2=make_pipeline(PolynomialFeatures(degree=2),LinearRegression())
model2.fit(x_train_transformed,y_train)
y_predict2=model2.predict(x_test_transformed)
print(r2_score(y_test,y_predict2))
```

```python
"""     Now making another model for prediting wicket to be 
         remain at last when inning will get over. """
```

```python
X=df[['batting_team','bowling_team','venue','pitch_condition','current_score','overs_completed','runs_last_5_overs','overs_remaining','run_rate','final_score']]
Y=df["wickets_remaining"]
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
preprocessor2= ColumnTransformer(transformers=[
    
    ('tnf1',OrdinalEncoder(categories=ordinal_order),['pitch_condition']),
    ('tnf2',OneHotEncoder(sparse_output=False,drop='first'),['batting_team','bowling_team','venue'])
],remainder='passthrough')
```

```python
# Transform
X_train_transformed = preprocessor2.fit_transform(X_train)
X_test_transformed = preprocessor2.transform(X_test)

# Convert back to DataFrame
X_train_transformed = pd.DataFrame(
    X_train_transformed,
    columns=preprocessor2.get_feature_names_out(),
    index=X_train.index
)

X_test_transformed = pd.DataFrame(
    X_test_transformed,
    columns=preprocessor2.get_feature_names_out(),
    index=X_test.index
)
```

```python
X_train_transformed
```

```python
model4=LinearRegression()
model4.fit(X_train_transformed,Y_train)
Y_predict=model4.predict(X_test_transformed)
print(r2_score(Y_test,Y_predict))
```

```python
model5=make_pipeline(PolynomialFeatures(degree=2),LinearRegression())
model5.fit(X_train_transformed,Y_train)
Y_predict2=model5.predict(X_test_transformed)
print(r2_score(Y_test,Y_predict2))
```

```python
model6=make_pipeline(PolynomialFeatures(degree=3),LinearRegression())
model6.fit(X_train_transformed,Y_train)
Y_predict2=model6.predict(X_test_transformed)
print(r2_score(Y_test,Y_predict2))
```
