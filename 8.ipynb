```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import OneHotEncoder
```

```python
df=pd.read_csv("student_performance.csv")
df.head()
```

```python
encoder=OneHotEncoder(sparse_output=False)
df['grade']=encoder.fit_transform(df[['grade']])
```

```python
df.head()
```

```python
df.drop(columns='student_id',inplace=True)
model=KMeans(n_clusters=3,random_state=0)
# 3 clusters for best,avg abd poor
model.fit(df)
result=model.labels_
df['cluster_performance']=result
df
```

```python
print(df['cluster_performance'].value_counts())
# here we can see our dataset divided into 3 categories
```

```python
cluster0_data = df[df['cluster_performance'] == 0]

print(cluster0_data)
#Seing only particular clustered data
```
