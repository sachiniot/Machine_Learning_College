```python
"""    Simple Linear Regression Model  """
```

```python

from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# giving Data
hours=[[1],[2],[3],[4],[5]]
marks=[40,50,65,70,80]
```

```python
# first plot the data if it look like linear data then use simple linear regression model 
plt.scatter(hours,marks,color="blue")  # data looks like mostly straight line so using LinearRegression model
```

```python
plt.plot(hours,marks)
# this also suggest mostly linear relation or straight line
```

```python
# making a object "model" of class "LinearRegression"
model=LinearRegression()
# Now traing and test the data for the model
model.fit(hours,marks)
# this will train on data means it will try to find the relatioon first if it finds linear then find the value of slope and intercept to fit in the equations.
```

```python
# print the value of the coef and intercept which model finds after training
print(model.intercept_)
```

```python
print(model.coef_)
```

```python
 # now predicting the value of scores after giving the value of the hours
predict_scores=model.predict([[6]])
print(predict_scores)
```
