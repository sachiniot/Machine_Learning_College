```python
"""  Now taking data of simple Linear regression from a dataset from kaggle: salary vs experience dataset  """

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
z=pd.read_csv("C:\\Users\\mukes\\OneDrive\\Documents\\ML\\Experience-Salary.csv")
z
```

```python
x = z[["exp(in months)"]]
y = z[["salary(in thousands)"]]
plt.plot(x,y, color="red")
# this tells about the linear data 
```

```python
model=LinearRegression()
model.fit(x,y)
print("predicted salary for 16.42 months: ",model.predict([[16.42]]))
```
